```python
def thaiweb(url):
    response = requests.get(url)  
    soup = BeautifulSoup(response.content, 'html.parser')
    meta_tag = soup.find('meta', property='og:title')

    # Extract the content attribute value if the tag exists
    if meta_tag:
        og_description = meta_tag['content']
        print(og_description)
    else:
        print("Meta tag with property='og:description' not found.")
        og_description ="None"
    important_info = soup.find('meta', property='og:description')

    # Extract the content attribute value if the tag exists
    if important_info:
        important = important_info['content']
        print(important)
    else:
        print("important_info was not found.")
        important="None"

    reseach_info = soup.find('div', class_='tabcontent news_desc', id='tap1')

    # Check if the <div> was found and extract text
    if reseach_info:
        reseach= reseach_info.get_text(separator="\n", strip=True)
        print(reseach)
    else:
        print("reseach_info was not found.")
        reseach="None"
    refer_info= soup.find('div', class_='tabcontent news_desc', id='tap4')

    # Check if the <div> was found and extract text
    if refer_info:
        refer = refer_info.get_text(separator="\n", strip=True)
        print(refer)
    else:
        print("Trefer_info was not found.")
        refer="None"
    location = soup.find('td', class_='toppics', string=lambda text: 'ที่ตั้ง :' in text if text else False)

    # Check if the <td> was found and extract the text from the adjacent <td>
    if location :
        td_value = location.find_next_sibling('td', style='font-family:fontThaipridi;')
        if td_value:
            location_text = td_value.get_text(strip=True)
            print(location_text)
        else:
            print("location was not found.")
    else:
        print("location was not found.")
        location_text = "None"
    coordinates_info= soup.find('td', class_='toppics', string=lambda text: 'พิกัด :' in text if text else False)

    # Check if the <td> was found and extract the text from the adjacent <td>
    if coordinates_info:
        td_value = coordinates_info.find_next_sibling('td', style='font-family:fontThaipridi;')
        if td_value:
            coordinates = td_value.get_text(strip=True)
            print(coordinates)
        else:
            print("The adjacent <td> with the specified style was not found.")
            coordinates="None"
    else:
        print("The <td> with the text 'ที่ตั้ง :' was not found.")
        
    age_info= soup.find('td', class_='toppics', string=lambda text: 'อายุสมัย :' in text if text else False)

    # Check if the <td> was found and extract the text from the adjacent <td>
    if age_info:
        td_value = age_info.find_next_sibling('td', style=lambda value: value and 'font-family:fontThaipridi' in value)
        if td_value:
            age_text = td_value.get_text(strip=True)
            print(age_text)
        else:
            print("The adjacent <td> with the specified style was not found.")
    else:
        age_text="None"
        print("The <td> with the text 'อายุสมัย :' was not found.")
    river_info = soup.find('td', class_='toppics', string=lambda text: 'แหล่งน้ำสำคัญ :' in text if text else False)

    # Check if the <td> was found and extract the text from the adjacent <td>
    if river_info:
        td_value = river_info.find_next_sibling('td', style='font-family:fontThaipridi;')
        if td_value:
            river = td_value.get_text(strip=True)
            print(river)
        else:
            print("แหล่งน้ำสำคัญ was not found.")
    else:
        print("แหล่งน้ำสำคัญ was not found.")
        river="None"
    h3_tag = soup.find('h3', class_='text-main text-bold mb-0', string='วันที่เผยแพร่ข้อมูล')

    # Check if the <h3> was found and extract the text from the adjacent <span>
    if h3_tag:
        span_tag = h3_tag.find_next('span', class_='text-bold text-secondary fs-18')
        if span_tag:
            date_text = span_tag.get_text(strip=True)
            print(date_text)
        else:
            print("The adjacent <span> with the specified class was not found.")
    else:
        print("The <h3> with the text 'วันที่เผยแพร่ข้อมูล' was not found.")
        date_text="None"
    tag_info = soup.find('td', class_='toppics', string=lambda text: 'คำสำคัญ/ป้ายกำกับ' in text if text else False)

    # Check if the <td> was found and extract the text from the adjacent <td>
    if tag_info:
        td_value = tag_info.find_next_sibling('td', style=lambda value: value and 'font-family:fontThaipridi' in value)
        if td_value:
            # Extract text from all <span> elements within the <td>
            spans = td_value.find_all('span', class_='badge bg-secondary text-light')
            tags = [span.get_text(strip=True) for span in spans]
            print(tags)
        else:
            print("The adjacent <td> with the specified style was not found.")
    else:
        print("The <td> with the text 'คำสำคัญ/ป้ายกำกับ :' was not found.")
        tags="None"
    link_info = soup.find('meta', property='og:url')

# Check if the <meta> was found and extract the content attribute
    if link_info:
        link= link_info.get('content')
        print(link)
    else:
        print("link was not found.")
        link="None"

    
    return og_description ,important,reseach,refer,location_text,age_text,river,date_text,tags,link
```
